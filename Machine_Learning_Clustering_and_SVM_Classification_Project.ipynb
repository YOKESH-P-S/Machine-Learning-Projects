```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
import random
dtfiles = "Data.txt"

m = np.loadtxt(dtfiles, dtype=float, encoding='utf-8-sig')

#First segment-For K-mean clustering

x = m[:, 0]
y = m[:, 1]
c = m[:, 2]

x1_mean = np.mean([x[:200]])
y1_mean = np.mean([y[:200]])
x2_mean = np.mean([x[201:400]])
y2_mean = np.mean([y[201:400]])

print("Average of Cluster-1 data points")
print(x1_mean,y1_mean)
print("Average of Cluster-2 data points")
print(x2_mean,y2_mean)

x1 = [2]
y1 = [-1]
x2 = [0]
y2 = [0]

'''
random_indices = random.sample(range(400), 300)
print(random_indices)
'''

random_indices=[45, 110, 274, 215, 320, 46, 33, 96, 289, 26, 287, 170, 118, 53, 300, 252, 324, 34, 335, 8, 163, 99, 243, 172, 185, 78, 253, 399, 11, 388, 375, 264, 297, 254, 51, 241, 210, 31, 226, 339, 222, 5, 95, 93, 354, 28, 383, 209, 326, 312, 114, 338, 116, 286, 221, 80, 16, 307, 38, 141, 360, 292, 333, 88, 276, 86, 316, 329, 71, 364, 217, 89, 273, 228, 107, 231, 65, 183, 205, 124, 211, 153, 92, 381, 265, 24, 376, 19, 361, 261, 296, 302, 66, 349, 270, 131, 233, 140, 148, 314, 248, 345, 129, 246, 285, 260, 18, 306, 104, 249, 258, 342, 58, 55, 232, 21, 35, 223, 142, 27, 394, 15, 175, 310, 385, 337, 397, 389, 13, 64, 293, 367, 366, 155, 122, 188, 305, 384, 167, 135, 123, 144, 67, 225, 369, 229, 186, 69, 190, 125, 357, 42, 370, 102, 374, 343, 353, 97, 29, 332, 327, 278, 378, 239, 108, 82, 87, 164, 145, 150, 321, 54, 328, 245, 380, 39, 72, 111, 73, 161, 160, 263, 109, 323, 43, 391, 299, 181, 377, 106, 230, 219, 396, 40, 48, 121, 119, 173, 4, 22, 59, 214, 177, 70, 238, 151, 373, 63, 23, 346, 32, 105, 126, 84, 334, 74, 220, 379, 255, 94, 37, 10, 120, 315, 224, 146, 290, 352, 57, 271, 91, 6, 2, 98, 267, 204, 20, 336, 266, 208, 237, 115, 250, 386, 47, 207, 192, 280, 200, 392, 227, 269, 157, 303, 49, 311, 194, 313, 272, 341, 202, 156, 351, 294, 277, 133, 365, 257, 191, 212, 187, 30, 251, 68, 168, 348, 235, 152, 60, 79, 184, 198, 355, 138, 358, 363, 61, 234, 3, 298, 256, 309, 268, 275, 100, 344, 101, 139, 240, 395]
x_subset = x[random_indices]
y_subset = y[random_indices]

tolerance = 0.00001
while True:
    b1 = []
    b2 = []
    for i in range(300):
        j1 = np.sqrt((x_subset[i] - x1[0]) ** 2 + (y_subset[i] - y1[0]) ** 2)
        j2 = np.sqrt((x_subset[i] - x2[0]) ** 2 + (y_subset[i] - y2[0]) ** 2)
        if j2 < j1:
            b2.append([x_subset[i], y_subset[i]])
        else:
            b1.append([x_subset[i], y_subset[i]])

    x1_new = np.mean([point[0] for point in b1])
    y1_new = np.mean([point[1] for point in b1])
    x2_new = np.mean([point[0] for point in b2])
    y2_new = np.mean([point[1] for point in b2])

    if abs(x1_new - x1[0]) < tolerance and abs(y1_new - y1[0]) < tolerance and \
       abs(x2_new - x2[0]) < tolerance and abs(y2_new - y2[0]) < tolerance:
        break

    x1[0] = x1_new
    y1[0] = y1_new
    x2[0] = x2_new
    y2[0] = y2_new

print("Updated Cluster 1 Center:", x1[0], y1[0])
print("Updated Cluster 2 Center:", x2[0], y2[0])

c1 = []
c2 = []
for i in range(len(random_indices)):
    h1 = np.sqrt((x[random_indices[i]] - x1[0]) ** 2 + (y[random_indices[i]] - y1[0]) ** 2)
    h2 = np.sqrt((x[random_indices[i]] - x2[0]) ** 2 + (y[random_indices[i]] - y2[0]) ** 2)
    if h1 < h2:
        c1.append([x[random_indices[i]], y[random_indices[i]], random_indices[i]])
    else:
        c2.append([x[random_indices[i]], y[random_indices[i]], random_indices[i]])

mis = []

for i in range(len(c1)):
    if c1[i][2]>200:
        mis.append([c1[i][0], c1[i][1], c1[i][2]])

for i in range(len(c2)):
    if c2[i][2]<200:
        mis.append([c2[i][0], c2[i][1], c2[i][2]])
print("Misclassified data points of randomly selected 300  points")
print(mis)
misc= np.array(mis)

left=[]

for i in range(400):
  if i not in random_indices:
    left.append(i)

c11=[]
c21=[]
for i in range(len(left)):
  h1 = np.sqrt((x[left[i]] - x1[0]) ** 2 + (y[left[i]] - y1[0]) ** 2)
  h2 = np.sqrt((x[left[i]] - x2[0]) ** 2 + (y[left[i]] - y2[0]) ** 2)
  if h1<h2:
    c11.append([x[left[i]],y[left[i]],left[i]])
  else:
    c21.append([x[left[i]],y[left[i]],left[i]])

mis1 = []

for i in range(len(c11)):
    if c11[i][2]>200:
        mis1.append([c11[i][0], c11[i][1], c11[i][2]])

for i in range(len(c21)):
    if c21[i][2]<200:
        mis1.append([c21[i][0], c21[i][1], c21[i][2]])
print("Misclassified data points of the left 100  points")
print(mis1)
misc1= np.array(mis1)



plt.figure(figsize=(10, 6))
plt.scatter([point[0] for point in b1], [point[1] for point in b1], c='b', marker='o', label='Cluster 1 Points')
plt.scatter([x1[0]], [y1[0]], c='r', marker='x', label='Cluster 1 Center')

plt.scatter([point[0] for point in b2], [point[1] for point in b2], c='r', marker='o', label='Cluster 2 Points')
plt.scatter([x2[0]], [y2[0]], c='b', marker='x', label='Cluster 2 Center')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of randomly selected 300 points and clustering of those points')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))

plt.scatter(x_subset, y_subset, c='c', marker='o', label='Data Points')
plt.scatter(misc[:, 0], misc[:, 1], marker='o', label='Miaclassified Data Points')
plt.scatter(x1_mean, y1_mean, c='r', marker='x', label='Avg of cluster point-1')
plt.scatter(x2_mean, y2_mean, c='g', marker='x', label='Avg of cluster point-2')
plt.scatter(x1[0], y1[0], c='b', marker='x', label='Cluster 1 Center')
plt.scatter(x2[0], y2[0], c='y', marker='x', label='Cluster 2 Center')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of randomly selected 300 points and misclassified data points')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))

plt.scatter([x1[0]], [y1[0]], c='b', marker='x', label='Cluster 1 Center')
plt.scatter([point[0] for point in c11], [point[1] for point in c11], c='b', marker='o', label='Cluster 1 Points')

plt.scatter([x2[0]], [y2[0]], c='r', marker='x', label='Cluster 2 Center')
plt.scatter([point[0] for point in c21], [point[1] for point in c21], c='r', marker='o', label='Cluster 2 Points')

plt.scatter(misc1[:, 0], misc1[:, 1], c='y',  marker='o', label='Miaclassified Data Points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of remaining 100 points and clustering of 100 points')
plt.grid(True)
plt.legend()
plt.show()



X = np.column_stack((x_subset, y_subset))

clf = svm.SVC(kernel='linear', C=1e6)

clf.fit(X, c[random_indices])

coef = clf.coef_[0]
intercept = clf.intercept_

plt.figure(figsize=(10, 6))
plt.scatter(x_subset, y_subset, c='c', marker='o', label='Data Points')
plt.scatter(x1, y1, c='r', marker='x', label='Cluster 1 Center')
plt.scatter(x2, y2, c='g', marker='x', label='Cluster 2 Center')

xx = np.linspace(min(x_subset), max(x_subset))
yy = -coef[0] / coef[1] * xx - intercept / coef[1]

plt.plot(xx, yy, label='Hyperplane')

margin = 1 / np.sqrt(np.sum(coef ** 2))
yy = yy + margin
plt.plot(xx, yy, 'k--', label='Margin')
yy = yy - 2 * margin
plt.plot(xx, yy, 'k--')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Support Vector Machine (SVM) for randomly selected 300 points')
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10, 6))

plt.scatter([point[0] for point in c11], [point[1] for point in c11], c='b', marker='o')

plt.scatter([point[0] for point in c21], [point[1] for point in c21], c='b', marker='o', label='Data Points of left 100 points')
xx = np.linspace(min(x_subset), max(x_subset))
yy = -coef[0] / coef[1] * xx - intercept / coef[1]

plt.plot(xx, yy, label='Hyperplane')

margin = 1 / np.sqrt(np.sum(coef ** 2))
yy = yy + margin
plt.plot(xx, yy, 'k--', label='Margin')
yy = yy - 2 * margin
plt.plot(xx, yy, 'k--')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Data points classification for left 100 points using SVM')
plt.grid(True)
plt.legend()
plt.show()


points_between_support_lines_300 = []

for i in random_indices:

    distance = abs(coef[0] * x[i] + coef[1] * y[i] + intercept) / np.sqrt(np.sum(coef ** 2))

    if 0 < distance < margin:
        points_between_support_lines_300.append([x[i], y[i], i])

print("Points between support lines for 300 points:")
for point in points_between_support_lines_300:
    print(f"Point: {point[:2]}, Index: {point[2]}")

points_between_support_lines_100 = []

for i in left:

    distance = abs(coef[0] * x[i] + coef[1] * y[i] + intercept) / np.sqrt(np.sum(coef ** 2))

    if 0 < distance < margin:
        points_between_support_lines_100.append([x[i], y[i], i])

print("Points between support lines for 100 points:")
for point in points_between_support_lines_100:
    print(f"Point: {point[:2]}, Index: {point[2]}")


a, b = coef
c = intercept
m=a/b
ic=c/b
print(f"Hyperplane Equation: y= {m}x+{ic}")
print("margin value:", margin)
print("points mispalced from cluster-1 for all 400 points")
for i in range(400):
  if i<200 and a*x[i]+b*y[i]+c>0:
    print(x[i],y[i],i)

print("points mispalced from cluster-2 for all 400 points ")
for i in range(400):
  if i>200 and a*x[i]+b*y[i]+c<0:
    print(x[i],y[i],i)


```
